# Fase 1: **API Advanced Search**

Come descritto nel report la fase di acquisizione dati è eseguita in due fasi diverse, una consequenziale all’altra. Nel primo step si sfruttano le funzionalità dell’**API Advanced Search** reperibile al sito: https://imdbapi.com/api#AdvancedSearch-header. 

Tramite l’API Advanced Search è possibile scaricare un insieme di film definendo diversi parametri. L’obiettivo del primo step è quello di ottenere un insieme di film popolari su IMDB dal 2010 al 2021. Si sceglie di scaricare 100 film (feature film) per ogni mese in base al punteggio di popolarità del film, di conseguenza si selezionano i 100 film più popolari in ordine decrescente (totale 1200 film per anno). 

Per acquisire dati si sfruttano le funzionalità della libreria python *requests*.

```python
import requests
```

```python
# Si definisce una lista con i mesi in cui si vogliono acquisire dati: da gennaio (01) a dicembre (12)
lista_mese = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
```

Si ricorda che con una chiave API si possono effettuare 100 richieste al giorno.

**Calcolo numero richieste da fare**: <br>
1 API di tipo Advanced Search con parametro il periodo temporale (1 mese) in cui si vogliono acquisire dati equivale ad una richiesta. <br>
Per 1 anno si devono effettuare 12 richieste. <br>
Per 12 anni (dal 2010 al 2021) si devono effettuare 144 richieste. <br>
Si creano 2 *lista_anno* diverse: la prima dal 2010 al 2017 (8 anni) e la seconda dal 2018 al 2021 (4 anni). Con la prima lista si faranno 96 richieste (12x8) e con la seconda lista si faranno 48 richieste (12x4).

```python
# Si definisce una lista con gli anni dal 2010 al 2017
lista_anno = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
```

```python
# Si definisce una lista con gli anni dal 2018 al 2021
lista_anno2 = ['2018', '2019', '2020', '2021']
```

Nel codice sottostante si acquisiscono dati da gennaio a dicembre per il periodo temporale dal 2010 al 2021.

**Nota**: I dati si acquisiscono in due parti:

1. **Parte 1** <br>
Acquisizione dati da gennaio a dicembre **dello stesso anno**: totale acquisizione su 11 mesi. Totale film per anno 1100. <br>
Ex: periodo temporale da 01-2010 a 12-2010
2. **Parte 2** <br>
Acquisizione dati da dicembre a gennaio **di due anni, uno consecutivo all'altro**: totale acquisizione su 1 mese. Totale film per anno 100. <br>
Ex: periodo temporale da 12-2020 a 01-2011

## Parte 1

I dati si salvano in un dizionario: la chiave è l'anno e il valore è una lista contenente 1200 film. In totale il dizionario ha 12 chiavi.

```python
# lista_mese = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
# lista_anno = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']

i=0
m=0
max = len(lista_mese)-1 #11 (deve arrivare fino alla posizione 10 perchè deve prendere il periodo temporale 11-12)
lista = []
diz = {}
while m < len(lista_anno) : 
    while i < max : 
        a2 = "https://imdb-api.com/API/AdvancedSearch/k_xl3l2xng?title_type=feature&release_date=" + lista_anno[m] + "-" + lista_mese[i] + "," + lista_anno[m] + "-" + lista_mese[i+1] + "&countries=us&sort=moviemeter,desc&count=100" 
        aa2 = requests.get(a2)
        aaa2 = aa2.json()
        lista.append(aaa2)
        i+=1
    i=0
    anno = lista_anno[m]
    diz[anno] = lista
    lista=[]
    m+=1
```

```python
# lista_mese = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
# lista_anno2 = ['2018', '2019', '2020', '2021']

i=0
m=0
max = len(lista_mese)-1 #11 (deve arrivare fino alla posizione 10 perchè deve prendere il periodo temporale 11-12)
lista2 = []
diz2 = {}
while m < len(lista_anno2) : 
    while i < max : 
        a3 = "https://imdb-api.com/API/AdvancedSearch/k_g105u00s?title_type=feature&release_date=" + lista_anno2[m] + "-" + lista_mese[i] + "," + lista_anno2[m] + "-" + lista_mese[i+1] + "&countries=us&sort=moviemeter,desc&count=100" 
        aa3 = requests.get(a3)
        aaa3 = aa3.json()
        lista2.append(aaa3)
        i+=1
    i=0
    anno = lista_anno2[m]
    diz2[anno] = lista2
    lista2=[]
    m+=1
```

Si uniscono i due dataset di output in un unico dataset chiamato *dizTot*.

```python
dizTot = {**diz, **diz2}
```

# Notebook 2: **API Title**

Come descritto nel report la fase di acquisizione dati è eseguita in due fasi diverse, una consequenziale all’altra. Nel secondo step si sfruttano le funzionalità dell’**API Title** reperibile al sito: https://imdb-api.com/api#Title-header. 

Ogni film è descritto univocamente da un'identificativo chiamato *id*. Gli id dei film che sono stati acquisiti con l'API Advanced Search sono immagazzinati in un file testo .txt in locale chiamato *LISTAID*.

Si ricorda che gli id ottenuti dalla fase precedente sono 14365. Inoltre, si ricorda che **ogni API Key può essere utilizzata per 100 richieste al giorno**. Per ottenere informazioni su tutti i 14365 film con una sola API Key ci vorrebbero 144 giorni. Di conseguenza, per riuscire a scaricare i dati in un tempo ragionevole si sono creati *23 account*, ottenendo così *23 API Key*. Gli account su IMDB sono stati creati utilizzando mail di amici e famigliari. Con 23 chiavi si possono fare *2300 richieste al giorno*. Di conseguenza, i dati sono scaricabili in *7 giorni*.

```python
# Importazione LISTAID e creazione del film mylist
my_file = open("C:/Users/user/Desktop/università DS/LISTAID.txt", "r")
mylist = my_file.read()
```

```python
mylist = mylist.split('\n')
```

```python
len(mylist)
```

Si creano 2 liste diverse:

1. *mylist_min*: lista dei primi 14300 id 
2. *mylist_partefin*: lista degli ultimi 65 film

```python
mylist_min = mylist[0:14300]
```

```python
len(mylist_min)
```

```python
mylist_partefin = mylist[14300:]
```

```python
len(mylist_partefin)
```

Considerando che ogni API Key può eseguire 100 richieste al giorno, si divide *mylist_min* in blocchi da 100. <br>
Il risultato è una lista di liste: la lista è composta da 143 elementi (143 blocchi), dove ogni blocco è composto da 100 elementi (100 id). <br>
Si crea *lista_fin*.

```python
listafin=[]
i=0

while i < len(mylist_min):
    lista = mylist[i:i+100]
    listafin.append(lista)
    i+=100
```

```python
len(listafin)
```

## Creazione database e collezione su **MongoDB**

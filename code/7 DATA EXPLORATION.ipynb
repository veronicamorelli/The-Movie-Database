## Data Exploration su **MongoDB**

In questa sezione si eseguono alcune interrogazioni connettendosi a MongoDB.

```python
import pprint
import pymongo
import pandas as pd

client = pymongo.MongoClient("mongodb://localhost:27017")

db = client['dataman']
print(db) # successfully connected

collection = db['movies']
print(collection)
```

```python
record = collection.find_one({}) # si cerca un record
```

```python
print(record) 
```

```python
collection.count_documents({}) # numero totale di documenti
```

```python
collection.find_one({"imDbRating": {"$exists": True}}, sort=[("imDbRating", -1)])  # film con maggiore punteggio imDbRating : Fanny Pay con 9.4 
```

```python
collection.count_documents({ "imDbRating": {"$exists": True}}) # quanti film hanno punteggio imDbRating : 2217
```

```python
collection.find_one({"imDbRating": {"$exists": True}}, sort=[("imDbRating", 1)]) # film con minore punteggio imDbRating : Intrinsic Leverage con 1.2
```

```python
collection.count_documents({"boxOffice.budget($)": {"$exists":True}}) # quanti film hanno il budget : 6560
```

```python
collection.find_one({"boxOffice.budget($)": {"$exists": True}}, sort=[("boxOffice.budget($)", 1)])  # Film con budget più basso: House on the Edge of Hell con 1$
```

```python
collection.find_one({"boxOffice.budget($)": {"$exists": True}}, sort=[("boxOffice.budget($)", -1)])  # Film con budget più alto: The Game of Freedom con 100000000$
```

## Data Exploration: **Profiling**

```python
from pandas_profiling import ProfileReport
```

```python
all_records = collection.find({"boxOffice.budget($)":{"$exists":True}}) # all_records: tutti i film che hanno l'attributo budget($)
```

```python
list_cursor = list(all_records) # si trasforma all_records in lista e si crea lest_cursor
```

```python
df = pd.DataFrame(list_cursor) # si trasforma list_cursor in data e si crea df
```

```python
df.columns # variabili di df
```

```python
df2 = df[['runtimeMins', 'imDbRating']] #si selezionano le variabili di interesse per l'esplorazione
```

```python
profile = ProfileReport(df2, title="Pandas Profiling Report") 
```
